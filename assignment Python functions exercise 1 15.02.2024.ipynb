#### 2. Write a program to read project score and exam score from the keyboard and print a bar chart of the scores and a grade. Follow the example below:

Enter project score

30

Enter exam score

40

Total: 70

Project: ======

Exam ======

Total ========

Grade: A

Note how the colons (:) are aligned.

Note the extra blank lines in the expected output.

The bar chart has one for each 5 marks 24 marks should be shown as ====. Assume all inputs are positive integers.

Grading rules:

A: at least 25 marks for each component and at least 60 in total

B: at least one component has 25 or more marks and at least 50 in total

C: at least 40 in total

D: otherwise

```python
def enter_score(x):
    while True:
        try:
            score=int(input(x))
            if score >=0:
                return score
            else:
                print("Please enter a positive integer: ")
        except:
            print("invalid input. Please enter a positive integer: ")

def cal_grade(project_score,exam_score):
    total_score = project_score + exam_score
    if project_score >=25 and exam_score >=25 and total_score >= 60:
        return "A Grade"
    elif (project_score >=25 or exam_score >=25) and total_score >= 50:
        return "B Grade"
    elif total_score >=40:
        return "c Grade"
    else:
        return "D Grade"

def bar_chart(score):
    return "=" * (score//5)

def main():
    project_score = enter_score ("enter project score : ")
    exam_score = enter_score ("enter exam score : ")
    total_score =  project_score + exam_score
    grade =cal_grade(project_score,exam_score)

    print("total score: ",total_score)
    print("project score: ",bar_chart(project_score))
    print("exam score: ",bar_chart(exam_score))
    print("total : ",bar_chart(total_score))
    print("grade : ", grade)
main()
```
